```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
happiness=pd.read_csv('C:/Users/D17297/Downloads/0.05a65o3cwn9t0.b53bayxyqtvhappiness_report.csv')
```

```python
happiness
```

```python
Mexico = happiness[(happiness['Country or region']=='Mexico')]
Mexico.set_index('Country or region',inplace=True)
Mexico
```

```python
plt.figure(figsize=(11,7))
plt.hist(happiness['Score'],color='mediumaquamarine')
props_bbox={'boxstyle': 'circle', 'facecolor':'red', 'alpha':0.5}
plt.text(Mexico['Score'],0,'MEX',bbox=props_bbox, fontdict={'color':'white'},fontsize='xx-large')
plt.title('World Happines Report')
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(9,7))
happiness['Healthy life expectancy'].plot.box()
plt.title('World Happines Report')
props_bbox={'boxstyle': 'circle', 'facecolor':'red', 'alpha':0.3}
plt.annotate('MEX',xy=(1,Mexico['Healthy life expectancy']), xytext=(1.2,1),arrowprops=dict(arrowstyle='<->', connectionstyle='angle3'),bbox=props_bbox, color='white',fontsize='x-large')
```

```python
Mexico1=Mexico.drop(['Overall rank','Score'],axis=1).T
Mexico1.rename_axis("", axis="columns",inplace=True)
Mexico1.rename(columns={'Mexico':'Factors'},inplace=True)
Mexico1
```

```python
plt.figure(figsize=(12,12))
plt.title('Contribution in Mexico Happiness Score',fontdict={'fontsize':'xx-large'}, position=(.5,.95))
plt.pie('Factors',labels=Mexico1.index, data=Mexico1,autopct='%1.2f%%')
```

```python
happiness.sort_values('Score', inplace=True)
```

```python
happiness1=pd.concat([happiness[-1:],happiness[happiness['Country or region']=='Mexico'],happiness[happiness['Country or region']=='Norway'],happiness[happiness['Country or region']=='Central African Republic'],happiness[:1]])
happiness1.set_index('Country or region',inplace=True)
happiness1.rename_axis("", axis="index",inplace=True)
happiness1=happiness1.drop(['Overall rank','Score'],axis=1).T
happiness1
```

```python
happiness1.plot(kind='bar',figsize=(12,7))
plt.xticks(fontsize='x-large')
plt.title('World Happines Report')
```

```python
fig,axs = plt.subplots(2,3,figsize=(15,9))
happiness.plot(kind='scatter',x='Score', y='GDP per capita',c='green',ax=axs[0,0])
happiness.plot(kind='scatter',x='Score', y='Social support',c='green',ax=axs[0,1])
happiness.plot(kind='scatter',x='Score', y='Healthy life expectancy',c='green',ax=axs[0,2])
happiness.plot(kind='scatter',x='Score', y='Freedom to make life choices',c='green',ax=axs[1,0])
happiness.plot(kind='scatter',x='Score', y='Generosity',c='green',ax=axs[1,1])
happiness.plot(kind='scatter',x='Score', y='Perceptions of corruption',c='green',ax=axs[1,2])
fig.suptitle('Contribution in Happiness Score',position=(.5,.95),fontsize='xx-large')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(round(happiness.corr(),2),annot=True)
plt.xticks(fontsize='xx-large')
plt.yticks(fontsize='xx-large')
```
